```python
import tarfile
import os

# Path to the downloaded .tar.gz file
tar_path = "/content/model_highres_52.tar.gz"

# Extract to /content/
extract_dir = "/content/"
with tarfile.open(tar_path, "r:gz") as tar:
    tar.extractall(path=extract_dir)
    print(f"âœ… Extracted to: {extract_dir}")

# Find the .pt file
pt_path = None
for root, dirs, files in os.walk(extract_dir):
    for file in files:
        if file.endswith(".pt"):
            pt_path = os.path.join(root, file)
            print(f"âœ… Found model file: {pt_path}")
            break

if pt_path is None:
    print("âŒ No .pt file found.")
else:
    # Optional: Load using Ultralytics YOLO
    try:
        from ultralytics import YOLO
        model = YOLO(pt_path)
        print("âœ… Model loaded successfully.")
    except Exception as e:
        print(f"âš ï¸ Could not load model: {e}")
```
